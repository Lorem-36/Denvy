```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("SET_10.csv")

df.fillna(df.mean(numeric_only=True), inplace=True)

df.drop_duplicates(inplace=True)

numerical_cols = df.select_dtypes(include=['number']).columns
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])


def grade_math(score):
    if score < 0.5:
        return 'F'
    elif score < 0.6:
        return 'D'
    elif score < 0.7:
        return 'C'
    elif score < 0.8:
        return 'B'
    else:
        return 'A'

df['Math_Grade'] = df['Math'].apply(grade_math)


df['Math_Bin'] = pd.cut(df['Math'], bins=3, labels=["Low", "Medium", "High"])


df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("SET-11.csv")

for col in ['Math', 'Science']:
    df[col] = df[col].fillna(df[col].mean())
df.drop_duplicates(inplace=True)


numerical_cols = df.select_dtypes(include=['number']).columns
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

def science_category(score):
    if score < 0.49:
        return 'Poor'
    elif score < 0.69:
        return 'Average'
    elif score < 0.89:
        return 'Good'
    else:
        return 'Excellent'

df['Science_Level'] = df['Science'].apply(science_category)

df['Science_Bin'] = pd.cut(df['Science'], bins=3, labels=["Low", "Medium", "High"])
df.head()
```
